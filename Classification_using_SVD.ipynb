```python
import pywt

# Perform wavelet transform
wavelet_data = []
for i in range(len(new)):
    coeffs = pywt.dwt(new.iloc[i, :], 'db1')
    cA, cD = coeffs
    wavelet_data.append(cA)

wavelet_data = np.array(wavelet_data)

# Perform SVD on wavelet-transformed data
U, S, V = np.linalg.svd(wavelet_data)

# Project the time series onto the first two principal components
projected_data2 = np.dot(wavelet_data, V.T[:, :2])
```
